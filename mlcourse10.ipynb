```python
train_df=pd.read_csv('https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/flight_delays_train.csv')
test_df=pd.read_csv('https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/flight_delays_test.csv')
```

```python
from sklearn.model_selection import cross_val_score, RandomizedSearchCV
```

```python
train_df
```

```python
test_df.reset_index(inplace=True)
```

```python
test_df.drop('index',axis=1,inplace=True)
```

```python
full_df=pd.concat([test_df,train_df])
```

```python
full_df
```

```python
full_df.reset_index(inplace=True)
```

```python
full_df.drop('index',axis=1,inplace=True)
```

```python
y_train=full_df.dep_delayed_15min[100000:]
```

```python
y_train=y_train.map({'Y':1,'N':0})
```

```python
X=full_df.drop('dep_delayed_15min',axis=1)
```

```python
full_df
```

```python
X
```

```python
X_dummies = pd.get_dummies(X, columns=X.columns[X.dtypes == "object"])
```

```python
X_dummies
```

```python
X_interact = X.copy()
```

```python
cat_features = X.columns[X.dtypes == "object"]
for i, col1 in enumerate(cat_features):
    for j, col2 in enumerate(cat_features[i + 1 :]):
        X_interact[col1 + "_" + col2] = X_interact[col1] + "_" + X_interact[col2]
```

```python
X_dummies
```

```python
X_train=X_dummies[100000:]
```

```python
fit1=glm(random_state=17,C=0.1)
```

```python
search=GridSearchCV(fit1,{'C':[0.01,0.1,1,10]},scoring='roc_auc',n_jobs=-1)
```

```python
fit1.fit(X_train,y_train)
```

```python
X_test=X_dummies[:100000]
```

```python
pred=fit1.predict_proba(X_test)[:,1]
```

```python
pred
```

```python
pred_df=pd.DataFrame({'id':test_df.index,'dep_delayed_15min':pred})
```
